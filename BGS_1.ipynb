```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
## https://data.desi.lbl.gov/public/edr/vac/edr/lss/v2.0/LSScats/clustering/
```

```python
bgs_data_N = Table.read("BGS_ANY_N_clustering.dat.fits")

bgs_data_S = Table.read("BGS_ANY_S_clustering.dat.fits")
```

```python
ii = bgs_data_N ['ROSETTE_NUMBER']==3
plt.scatter(bgs_data_N['RA'][ii], bgs_data_N['DEC'][ii], s=0.1)
plt.axis('equal')
```

```python
_=plt.hist(bgs_data_N['Z'][ii])
```

```python
jj= (bgs_data_N['Z']>0.25)&(bgs_data_N['Z']<0.3)
plt.scatter(bgs_data_N['RA'][ii&jj], bgs_data_N['DEC'][ii&jj],s=5)
plt.axis('equal')
```

```python

rosettes = sorted(set(bgs_data_N['ROSETTE_NUMBER']))

for ros_num in rosettes:
    ii = bgs_data_N['ROSETTE_NUMBER'] == ros_num
    jj = (bgs_data_N['Z'] > 0.25) & (bgs_data_N['Z'] < 0.3)

    # Plot 1: RA vs DEC (all)
    plt.figure(figsize=(6, 4))
    plt.scatter(bgs_data_N['RA'][ii], bgs_data_N['DEC'][ii], s=0.1)
    plt.title(f"Rosette {ros_num}: RA vs DEC (all)")
    plt.axis('equal')
    plt.xlabel('RA')
    plt.ylabel('DEC')
    plt.show()

    # Plot 2: Histogram of Z
    plt.figure(figsize=(6, 4))
    _ = plt.hist(bgs_data_N['Z'][ii], bins=50, color='gray')
    plt.title(f"Rosette {ros_num}: Histogram of Z")
    plt.xlabel('Z')
    plt.ylabel('Count')
    plt.show()

    # Plot 3: RA vs DEC (Z in 0.25–0.3)
    plt.figure(figsize=(6, 4))
    plt.scatter(bgs_data_N['RA'][ii & jj], bgs_data_N['DEC'][ii & jj], s=5, color='red')
    plt.title(f"Rosette {ros_num}: RA vs DEC (Z between 0.25–0.3)")
    plt.axis('equal')
    plt.xlabel('RA')
    plt.ylabel('DEC')
    plt.show()
```
